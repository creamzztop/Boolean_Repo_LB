# VERIFICA COLONNE DATI ISTAT

```python
import pandas as pd
```

```python
df_istat = pd.read_csv("../raw_data/ISTAT_raw.csv")
print(df_istat.dtypes)
```

# DATAFLOW (str)
"IT1:41_983(1.0)", probabilmente inutile

```python
print(df_istat["DATAFLOW"].unique())
print("________________________________________________")
print(df_istat["DATAFLOW"].value_counts())
print("________________________________________________")
print("Valori nulli: ", df_istat["DATAFLOW"].isnull().sum())
print("________________________________________________")
```

# FREQ (str)
"A", probabilmente inutile


```python
print(df_istat["FREQ"].unique())
print("________________________________________________")
print(df_istat["FREQ"].value_counts())
print("________________________________________________")
print("Valori nulli: ", df_istat["FREQ"].isnull().sum())
print("________________________________________________")
```

# REF_AREA (int64)
valori da 1001 a 111107, indicano comuni ma da verificare tra gli anni cosa cambia

```python
print(df_istat["REF_AREA"].unique())
print("________________________________________________")
print(df_istat["REF_AREA"].value_counts())
print("________________________________________________")
print("Valori nulli: ", df_istat["REF_AREA"].isnull().sum())
print("________________________________________________")
```
