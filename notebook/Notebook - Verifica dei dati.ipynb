# VERIFICA COLONNE DATI SITUAS

```python
df_2024 = pd.read_csv("../raw_data/comuni_2024-12-31.csv", sep=";")
df_2026 = pd.read_csv("../raw_data/comuni_2026-06-21.csv", sep=";")
```

```python
print(df_2024.columns)
print("_______________________________________")
print(df_2026.columns)
```

```python
print(df_2024.dtypes)
print("_______________________________________")
print(df_2026.dtypes)
```

# DIFFERENZE TRA 2026 E 2024

```python
print("comuni presenti 2024: ", len(df_2024))
print("comuni presenti 2026: ", len(df_2026))
print("differenza numerica: ", len(df_2026)-len(df_2024), " comuni nel 2026")
```

Considerando il periodo dello storico dati che si ferma al 2024 ma volendo evitare di fare operazioni su comuni che non esistono pie' trovo i comuni in comune tra il 2024 e oggi con inner join

```python
df_comuni_2426 = df_2024.merge(df_2026, how="inner", on="Codice Comune (numerico)").reset_index(drop=True)

print("Comuni presenti sia nel 2024 che nel 2026: ", len(df_comuni_2426))
```

```python
print(df_comuni_2426.shape)
print("________________________________")
print(df_comuni_2426.head(10))
```

```python
print(df_comuni_2426.columns)
```

```python
df_2426_clean = df_comuni_2426[[
    "Codice Ripartizione geografica_x",
    "Codice Regione_x",
    "Codice Provincia/Uts_x",
    "Codice Comune (alfanumerico)_x", 
    "Codice Comune (numerico)",
    "Comune_x",
    "Sigla automobilistica_x",
    "Capoluogo di Provincia/Uts_x",
    "Capoluogo di Regione_x",
    "Popolazione legale_x",
    "Anno Censimento_x",
    "Superficie (Kmq)_x",
    "Anno (Superficie)_x",
    "Popolazione residente_x",
    "Anno (Popolazione residente)_x"
    ]].copy()
```

```python
df_2426_clean.columns = [
    "ID_Ripartizione_Geografica",
    "ID_Regione",
    "ID_Provincia/Uts",
    "IDA_Comune", 
    "IDN_Comune",
    "Comune",
    "Sigla_Auto",
    "IS_Capoluogo_Provincia/Uts",
    "IS_Capoluogo_Regione",
    "Pop_Legale",
    "Anno_Pop_Legale",
    "Superficie_Kmq",
    "Anno_Superficie",
    "Pop_Residente",
    "Anno_Pop_Residente"
]

print(df_2426_clean.dtypes)
df_2426_clean.head(3)
```

```python
df_2426_clean["Superficie_Kmq"] = df_2426_clean["Superficie_Kmq"].astype(str).str.replace(".", "", regex=False)
df_2426_clean["Superficie_Kmq"] = df_2426_clean["Superficie_Kmq"].astype(str).str.replace(",", ".", regex=False).astype(float)
print(df_2426_clean["Superficie_Kmq"].dtypes)
```
