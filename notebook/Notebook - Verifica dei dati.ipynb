Considerando il periodo dello storico dati che si ferma al 2024 ma volendo evitare di fare operazioni su comuni che non esistono pie' trovo i comuni in comune tra il 2024 e oggi con inner join

```python
df_comuni_2426 = df_2024.merge(df_2026, how="inner", on="Codice Comune (numerico)").reset_index(drop=True)

print("Comuni presenti sia nel 2024 che nel 2026: ", len(df_comuni_2426))
```

```python
df_2426_clean.loc[df_2426_clean["IDN_Comune"] == 1168, "Comune"] = "None"

print(df_2426_clean[df_2426_clean["IDN_Comune"] == 1168][["IDN_Comune", "Comune"]])
```

```python
count = {}

for nome in df_2426_clean["Comune"]:
    if nome in count:
        count[nome] += 1
    else:
        count[nome] = 1

duplicati = []

for nome, conteggio in count.items():
    if conteggio > 1:
        duplicati.append(nome)

print("Nomi duplicati trovati: ", len(duplicati))
print(duplicati)
print("_____________________________________________________")

for nome in duplicati:
    print(nome)
    for index, riga in df_2426_clean.iterrows():
        if riga["Comune"] == nome:
            print("IDN: ", riga["IDN_Comune"], "Provincia: ", riga["Sigla_Auto"], "Regione: ", riga["ID_Regione"])
```
