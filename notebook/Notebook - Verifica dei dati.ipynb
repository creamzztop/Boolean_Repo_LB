## ESPORTAZIONE COLONNE ISTAT

```python
df_incidenti = df_istat[df_istat["DATA_TYPE"] == "ROADACC"][["REF_AREA", "TIME_PERIOD", "OBS_VALUE"]]
df_feriti = df_istat[(df_istat["DATA_TYPE"] == "KILLINJ") & (df_istat["RESULT"] == "F")][["REF_AREA", "TIME_PERIOD", "OBS_VALUE"]]
df_morti = df_istat[(df_istat["DATA_TYPE"] == "KILLINJ") & (df_istat["RESULT"] == "M")][["REF_AREA", "TIME_PERIOD", "OBS_VALUE"]]
```

```python
df_incidenti.columns = [
    "IDN_Comune",
    "Anno",
    "n_Incidenti"].copy()

df_feriti.columns = [
    "IDN_Comune",
    "Anno",
    "n_Feriti"].copy()

df_morti.columns = [
    "IDN_Comune",
    "Anno",
    "n_Morti"].copy()
```

```python
print(df_incidenti.dtypes)
print("______________________")
print(df_feriti.dtypes)
print("______________________")
print(df_morti.dtypes)
print("______________________")
```

```python
df_istat_clean = df_incidenti.merge(df_feriti, on=["IDN_Comune", "Anno"], how="left").merge(df_morti, on=["IDN_Comune", "Anno"], how="left")
```

```python
print(df_istat_clean.shape)
print(df_istat_clean.columns.tolist())
df_istat_clean.head(3)
```

```python
df_istat_clean.to_csv("../clean_data/ISTAT_clean.csv", index=False)
```

```python
print(df_istat_clean.shape)
print(df_istat_clean.columns.tolist())
```

```python
count = {}

for nome in df_2426_clean["Comune"]:
    if nome in count:
        count[nome] += 1
    else:
        count[nome] = 1

duplicati = []

for nome, conteggio in count.items():
    if conteggio > 1:
        duplicati.append(nome)

print("Nomi duplicati trovati: ", len(duplicati))
print(duplicati)
print("_____________________________________________________")

for nome in duplicati:
    print(nome)
    for index, riga in df_2426_clean.iterrows():
        if riga["Comune"] == nome:
            print("IDN: ", riga["IDN_Comune"], "Provincia: ", riga["Sigla_Auto"], "Regione: ", riga["ID_Regione"])
```
